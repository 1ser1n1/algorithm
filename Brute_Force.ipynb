위 사진처럼 건물과 건물 사이에 얼만 큼 빗물이 담길지 계산하는 문제

함수`trapping_rain`은 건물의 높이에 대한 정보를 보관하는 리스트buildings를 파라미터로 받아 담기는 빗물의 총량을 리턴

```python
def trapping_rain(buildings):
  n = buildings[0] #맨 처음부터 인덱싱을 할 경우마다 가장 큰 건물의 높이
  rain_total = 0 #빗물의 합
  max_building = max(buildings) #가장 큰 건물의 높이
  max_building_index = buildings.index(max_building) #가장큰 건물의 인덱스번호
   
  for i in buildings[1:max_building_index]: #맨처음은 빗물이 고이지 않으므로 그다음부터 가장 큰 건물의 앞까지 인덱싱
    if i < n:
      rain_total += n-i #인덱싱한 건물이 그전의 건물의 높이보다 작을경우 차를 구해 빗물을 구함
    else: #인덱싱한 건물이 더 클경우
      n = i #가장 큰 건물을 바꿔줌

#반대로 뒤집어 맨 마지막부터 가장 큰 건물까지 구함
  n = buildings[-1] #맨 마지막부터 인덱싱을 할 경우, 현재 가장 큰 건물의 높이
  for i in buildings[-2:max_building_index:-1]: #맨마지막은 물이 고이지 않음
    if i < n:
      rain_total += n-i
    else:
      n = i
   
  return rain_total
# 테스트
print(trapping_rain([3, 0, 0, 2, 0, 4]))
print(trapping_rain([0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]))
```
